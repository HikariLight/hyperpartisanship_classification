```python
def read_numbers_from_file(filename):
    numbers = []
    with open(filename, 'r') as file:
        for line in file:
            # Strip whitespace and convert to integer
            number = line.strip()
            numbers.append(number)
    return numbers

# Usage
filename = 'cleanooutput.txt'  # Replace with your file name
number_list = read_numbers_from_file(filename)

print(len(number_list))
```

```python
import re

def extract_integers(input_list):
    # This will store the extracted integers
    integers = []
    
    # Define a regex pattern to match integers
    integer_pattern = re.compile(r'\b\d+\b')
            
    for index, item in enumerate(input_list):
        # Find all integers in the string item
        found_integers = integer_pattern.findall(item)
        
        if found_integers:
            # Convert found integers to int
            int_found_integers = list(map(int, found_integers))
           
            # Add the found integers to the result list
            integers.extend(int_found_integers)
            
            # Check if there are more than one integer and print the element if true
            if len(int_found_integers) > 1:
                print(f"Element at index {index} contains more than one integer: {item}")
        else:
            # Check for keywords if no integers are found
            item_lower = item.lower()
            if 'verdadera' in item_lower or 'veraz' in item_lower:
                integers.append(0)
            elif 'falsa' in item_lower:
                integers.append(1)
            else:
                # Print the index if neither integers nor keywords are found
                print(f"No integer or keyword found in element at index {index}")
                #integers.append(0)  # Append 0 if no integer or keyword is found
                integers.append(2)
    return integers
```

```python
labels_parsed = extract_integers(number_list)
```

```python
def compute_metrics(predictions, labels):
        accuracy_metric = evaluate.load("accuracy")
        precision_metric = evaluate.load("precision")
        recall_metric = evaluate.load("recall")
        f1_metric = evaluate.load("f1")


        accuracy = accuracy_metric.compute(predictions=predictions, references=labels)[
            "accuracy"
        ]
        precision = precision_metric.compute(
            predictions=predictions, references=labels, average="weighted"
        )["precision"]
        recall = recall_metric.compute(
            predictions=predictions, references=labels, average="weighted"
        )["recall"]
        f1 = f1_metric.compute(
            predictions=predictions, references=labels, average="weighted"
        )["f1"]

        return {
            "accuracy": accuracy,
            "precision": precision,
            "recall": recall,
            "f1-score": f1,
        }


results=compute_metrics(labels_parsed, gold_ans)
print(f"FNSpanish, CoT, prompt with temperature=0.1:\n", json.dumps(results, indent=4))
```
